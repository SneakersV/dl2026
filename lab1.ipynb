```python
# No numpy, no pandas, no GPT, window is the best

# Assume that the funcion f(x) = x^2
def gradient_descent(w, lr):
    return w - lr * (2 * w)
    
def show_table(times, lr, x):
    print("-" * 20)
    print("Time\tx\tf(x)")
    for i in range(times):
        f_x = x ** 2
        print(f"{i+1}\t{x:.2f}\t{f_x:.2f}")
        x = gradient_descent(x, lr)
```

```python
show_table(times=10, lr=0.1, x=8)
```
