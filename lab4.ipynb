```python
# Build nerual network from scratch

def cal_layer_output(l1, l2):
    output = []
    for j in range(len(l2)):
        sum = 0
        for i in range(len(l1)):
            sum += l1[i] * l2[j]
        output.append(sum)
    return output

class Neuron:
    def __init__(self, weight=0):
        self.weights = weight
    
    def set_weights(self, new_weights):
        self.weights = new_weights
        
    def get_weights(self):
        return self.weights
        
        
class Layer:
    def __init__(self, n_neurons, weights=[]):
        self.neurons = []
        if weights:
            if len(weights) <= n_neurons:
                for i in range(len(weights)):
                    print(f"creating neuron with weights: {weights[i]}")
                    neuron = Neuron(weights[i])
                    self.neurons.append(neuron)
            for i in range(len(weights), n_neurons):
                print(f"creating neuron with default weights: 0")
                neuron = Neuron()
                self.neurons.append(neuron)
                
        else:
            for i in range(n_neurons):
                print(f"creating neuron with default weights: 0")
                neuron = Neuron()
                self.neurons.append(neuron)
            
    def __len__(self):
        return len(self.neurons)
    
    def get_layer_weights(self):
        weights = []
        for neuron in self.neurons:
            weights.append(neuron.get_weights())
        return weights


class NeuralNetwork:
    def __init__(self, n_layers, n_neurons_per_layer, weights=[]):
        print(f"Creating NN with {n_layers} layers and {n_neurons_per_layer} neurons per layer")
        self.layers = []
        
        if weights:
            if len(weights) <= n_layers:
                for i in range(len(weights)):
                    print(f"Creating layer {i+1} with {n_neurons_per_layer[i]} neuron(s), weights: {weights[i]}")
                    layer = Layer(n_neurons_per_layer[i], weights=weights[i])
                    print(f"Layer {i+1} created with weights: {layer.get_layer_weights()}")
                    self.layers.append(layer)
                
                for i in range(len(weights), n_layers):
                    print(f"Creating layer {i+1} with {n_neurons_per_layer[i]} neuron(s), default weights: {[0]*n_neurons_per_layer[i]}")
                    layer = Layer(n_neurons_per_layer[i])
                    print(f"Layer {i+1} created with weights: {layer.get_layer_weights()}")
                    self.layers.append(layer)
        else:
            for i in range(n_layers):
                print(f"Creating layer {i+1} with {n_neurons_per_layer[i]} neuron(s), default weights: {[0]*n_neurons_per_layer[i]}")
                layer = Layer(n_neurons_per_layer[i])
                print(f"Layer {i+1} created with weights: {layer.get_layer_weights()}")
                self.layers.append(layer)
                
        self.input = self.layers[0]
    
    
    def __len__(self):
        return len(self.layers)
                
            
    def feed_forward(self):
        input_data = self.input.get_layer_weights()
        outputs = []
        for layer in self.layers[1:]:
            weights = layer.get_layer_weights()
            output = cal_layer_output(input_data, weights)
            outputs.append(output)
            input_data = output
        return outputs
    
    def get_params(self):
        params = []
        for layer in self.layers:
            params.append(layer.get_layer_weights())
        return params
```

```python
path = r".\nn.txt"
with open(path, "r") as f:
    data = f.read().split()

n_layers = int(data[0])
layers = []

print(f"The number of layers: {n_layers}")

for i in range(n_layers):
    layers.append(int(data[i+1]))
    print(f"Number of neurons in layer {i+1}: {layers[i]}")

print()
nn = NeuralNetwork(n_layers, layers, weights=[[1,2,3], [1,2,3,4]])
print()
print(f"Number of layers in the neural network (include input/output layer): {len(nn)}")
print(f"Parameters of the neural network: {nn.get_params()}")

output = nn.feed_forward()
print(f"Output of the neural network after feed forward: {output}")
```
